# Introduction to Probability

Probability is simply how likely an event is to happen.

- The probability of an event is between 0 and 1.
- The probability of a certain event is 1.
- The probability of an impossible event is 0.

## Probability in Terms of Data Science
In terms of data science, probability is how likely we can find a value within our data.


### Simple Terms in Probability
- **Experiment**: A process by which an observation is obtained is called an experiment.
- **Outcome**: An observed result of an experiment is called an outcome.
- **Event**: A set containing outcomes of an experiment is called an event.
- **Sample Space**: A set of all possible events of an experiment is called a sample space.
- **Probability of an Event**: It is the ratio of the number of ways an event can occur to the number of possible outcomes.



## Random Variable

#### A random variable is like a magic box that can give you different outcomes. These outcomes are the possible toys you can get.

## Types of Random Variables

### `Discrete Random Variable`

Imagine you have a dice with numbers 1 to 6. When you roll the dice, you can get any one of these numbers. This is a discrete random variable because you can count the possible outcomes (1, 2, 3, 4, 5, or 6).

### `Continuous Random Variable`

Now, imagine you have a magic ruler that can measure your height. Your height can be any number, like 100.5 cm, 100.55 cm, or even 100.555 cm. This is a continuous random variable because there are so many possible outcomes that you canâ€™t count them all.

## `Examples`
#### *`Discrete Random Variable`*: Rolling a dice or picking a card from a deck.
#### *`Continuous Random Variable`*: Measuring your height or the temperature outside.
So, a random variable is like a magic box that gives you different outcomes, and it can be either discrete (countable outcomes) or continuous (uncountable outcomes).

```python
## Discrete Random Varable

import numpy as np
import matplotlib.pyplot as plt

# Possible outcomes of rolling a six-sided die
outcomes = [1, 2, 3, 4, 5, 6]
# Probability of each outcome (since it's a fair die)
probabilities = [1/6] * 6

# Simulate rolling the die 1000 times
rolls = np.random.choice(outcomes, size=1000, p=probabilities)

# Plot the results
plt.hist(rolls, bins=np.arange(1, 8) - 0.5, edgecolor='black', density=True)
plt.xticks(outcomes)
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.title('Discrete Random Variable: Rolling a Six-Sided Die')
plt.show()
```

```python
## Continuous Random Variable

from scipy.stats import norm

# Generate data for heights (mean=170 cm, standard deviation=10 cm)
mean_height = 170
std_dev_height = 10
heights = np.random.normal(mean_height, std_dev_height, 1000)

# Plot the results
plt.hist(heights, bins=30, edgecolor='black', density=True)
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.title('Continuous Random Variable: Heights')
plt.show()
```

## `Probability Function`
A probability function is like a magic rule that tells us the chance of picking each color of marble from the bag. For example, if we have a bag with 3 red marbles and 2 blue marbles, the magic rule (probability function) tells us:

The chance of picking a red marble is 3 out of 5.
The chance of picking a blue marble is 2 out of 5.

## `Probability Distribution`
A probability distribution is like a picture or a list that shows all the possible outcomes and their chances. It uses the magic rule (probability function) to show the chances for all the marbles in the bag.

For our marbles, the probability distribution looks like this:

Red marble: 3/5 or 60%
Blue marble: 2/5 or 40%
How They Are Related
The probability function gives us the rules, and the probability distribution shows us the whole picture using those rules.

```python
def probability_function(color, total_marbles, color_count):
    return color_count / total_marbles

# Total marbles
total_marbles = 5
# Number of red and blue marbles
red_marbles = 3
blue_marbles = 2

# Probability of picking a red or blue marble
prob_red = probability_function('red', total_marbles, red_marbles)
prob_blue = probability_function('blue', total_marbles, blue_marbles)

print(f"Probability of picking a red marble: {prob_red}")
print(f"Probability of picking a blue marble: {prob_blue}")
```

```python
import matplotlib.pyplot as plt

# Colors and their probabilities
colors = ['Red', 'Blue']
probabilities = [prob_red, prob_blue]

# Plot the probability distribution
plt.bar(colors, probabilities, color=['red', 'blue'])
plt.xlabel('Color of Marble')
plt.ylabel('Probability')
plt.title('Probability Distribution of Marbles')
plt.show()
```

## `Probability Distribution Function`: Specific type of probability function that includes PMF, PDF, and CDF, describing how probabilities are distributed over the values of a random variable.


## Types:

#### `Probability Mass Function (PMF)`:
 This is like a map showing where most candies are in the jar. It tells us the likelihood of picking a candy within a certain range. Used for discrete random variables.

#### `Probability Density Function (PDF)`: 
This is for counting candies. It tells us the exact probability of picking a specific candy.used for continuous random variables.

#### `Cumulative Distribution Function (CDF)`:
 This is like a storybook that tells us the chances of picking a candy up to a certain point. It adds up all the probabilities from the start to that point. Used for both discrete and continuous random variables.

```python
## PMF (Probability Mass Function)

import matplotlib.pyplot as plt

# Possible outcomes
outcomes = [1, 2, 3, 4, 5, 6]
# Probability of each outcome
probabilities = [1/6] * 6

plt.bar(outcomes, probabilities)
plt.xlabel('Die Face')
plt.ylabel('Probability')
plt.title('Probability Mass Function of Rolling a Die')
plt.show()
```

```python
## Probability Density Function (PDF)

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Generate data
mean = 170  # average height
std_dev = 10  # standard deviation
heights = np.linspace(140, 200, 1000)
probabilities = norm.pdf(heights, mean, std_dev)

plt.plot(heights, probabilities)
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Heights')
plt.show()
```

```python
## CDF (Cumulative Distribution Function)

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Generate data
mean = 170  # average height
std_dev = 10  # standard deviation
heights = np.linspace(140, 200, 1000)
cumulative_probabilities = norm.cdf(heights, mean, std_dev)

plt.plot(heights, cumulative_probabilities)
plt.xlabel('Height (cm)')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function of Heights')
plt.show()
```

# `Discrete Probability Distributions`


## `Bernoulli Distribution`
Explanation: Imagine flipping a coin. It can land on heads (1) or tails (0). The Bernoulli distribution shows the probability of getting heads or tails.

`Example`: Flipping a coin.

```python
import matplotlib.pyplot as plt
import numpy as np

# Probability of heads (success)
p = 0.5
# Possible outcomes
outcomes = [0, 1]
# Probabilities of outcomes
probabilities = [1-p, p]

plt.bar(outcomes, probabilities)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution')
plt.xticks(outcomes, ['Tails (0)', 'Heads (1)'])
plt.show()
```

## `Binomial Distribution`
Explanation: This is like flipping a coin multiple times and counting the number of heads. It shows the probability of getting a certain number of successes in a fixed number of trials.

`Example`: Flipping a coin 10 times.

```python
import matplotlib.pyplot as plt
from scipy.stats import binom

# Number of trials and probability of success
n, p = 10, 0.5
x = np.arange(0, n+1)
probabilities = binom.pmf(x, n, p)

plt.bar(x, probabilities)
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title('Binomial Distribution of Coin Flips')
plt.show()
```

## `Poisson Distribution`
Explanation: This is used for counting the number of events happening in a fixed interval of time or space, like the number of emails you get in an hour.

`Example`: Number of emails received in an hour

```python
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Average number of emails (lambda)
lambda_ = 3
x = np.arange(0, 15)
probabilities = poisson.pmf(x, lambda_)

plt.bar(x, probabilities)
plt.xlabel('Number of Emails')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.show()
```

# `Continuous Probability Distributions`

## `Normal Distribution`
Explanation: This is the famous bell curve. Most values are around the mean, with fewer values as you move away. Heights of people often follow this distribution.

`Example`: Heights of students.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Generate data
mean = 170  # average height
std_dev = 10  # standard deviation
heights = np.linspace(140, 200, 1000)
probabilities = norm.pdf(heights, mean, std_dev)

plt.plot(heights, probabilities)
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Heights')
plt.show()
```

# `Uniform Distribution`
Explanation: Every outcome in a range is equally likely. Itâ€™s like picking a random number between 0 and 1.

`Example`: Random numbers between 0 and 1

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
data = np.random.uniform(0, 1, 1000)

plt.hist(data, bins=30, density=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Uniform Distribution')
plt.show()
```

# `Log-Normal Distribution`
Explanation: This distribution is used for data that can grow exponentially, like stock prices.

`Example`: Simulating stock prices

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
mean = 0
sigma = 0.1
data = np.random.lognormal(mean, sigma, 1000)

plt.hist(data, bins=30, density=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Log-Normal Distribution')
plt.show()
```

# `Power Law Distribution`
Explanation: This distribution is used for data that follows a power law, like the distribution of wealth.

`Example`: Simulating wealth distribution

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
a = 3.5  # shape parameter
data = np.random.pareto(a, 1000)

plt.hist(data, bins=30, density=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Power Law Distribution')
plt.show()
```

## `Pareto Distribution`
Explanation: This is a specific type of power law distribution, often used to describe the distribution of wealth.

`Example`: Simulating wealth distribution

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
a = 3.0  # shape parameter
data = (np.random.pareto(a, 1000) + 1) * 1000

plt.hist(data, bins=30, density=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Pareto Distribution')
plt.show()
```
